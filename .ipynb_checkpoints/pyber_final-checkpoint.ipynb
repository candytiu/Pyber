```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
full_df = pd.merge(ride_data, city_data, how ="left", on="city")

# Display the data table for preview
full_df.head(10)
```

```python
# Urban information 
urban_data = full_df.loc[full_df['type']=='Urban']
urban_total_rides = urban_data.groupby(["city"])["ride_id"].count()
urban_avg_fare = round(urban_data.groupby(["city"])["fare"].mean(),2)
urban_driver_count = urban_data.groupby(["city"])["driver_count"].first()

# Suburban information 
suburban_data = full_df.loc[full_df['type']=='Suburban']
suburban_total_rides = suburban_data.groupby(["city"])["ride_id"].count()
suburban_avg_fare = round(suburban_data.groupby(["city"])["fare"].mean(),2)
suburban_driver_count = suburban_data.groupby(["city"])["driver_count"].first()
print(suburban_driver_count)
# Rural information 
rural_data = full_df.loc[full_df['type']=='Rural']
rural_total_rides = rural_data.groupby(["city"])["ride_id"].count()
rural_avg_fare = round(rural_data.groupby(["city"])["fare"].mean(),2)
rural_driver_count = rural_data.groupby(["city"])["driver_count"].first()

```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
## Urban x-axis and y-axis
urban_ride_count = [x1 for x1 in urban_total_rides]
avg_urban_fare = [y1 for y1 in urban_avg_fare]
## Suburban x-axis and y-axis
suburban_ride_count = [x2 for x2 in suburban_total_rides]
avg_suburban_fare = [y2 for y2 in suburban_avg_fare]
## Rural x-axis and y-axis
rural_ride_count = [x3 for x3 in rural_total_rides]
avg_rural_fare = [y3 for y3 in rural_avg_fare]

# Build the scatter plots for each city types
plt.scatter(urban_ride_count, avg_urban_fare, marker="o", facecolors="lightcoral", edgecolors="black",
            s=5*urban_driver_count, alpha=0.75,label="Urban")
plt.scatter(suburban_ride_count, avg_suburban_fare, marker="o", facecolors="lightskyblue", edgecolors="black",
            s=5*suburban_driver_count, alpha=0.75,label="Suburban")
plt.scatter(rural_ride_count, avg_rural_fare, marker="o", facecolors="gold", edgecolors="black",
            s=5*rural_driver_count, alpha=0.75,label="Rural")
# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data(2016)")
plt.xlabel("Total number of rides (Per City)")
plt.ylabel("Average fare ($)")
plt.grid(b=None,which='major',axis='both',color='grey')
# Create a legend
plt.legend(loc="best",title="City Type")

# Incorporate a text label regarding circle size
textstr = "Note: \nCircle size correlates with driver count per city"
plt.gcf().text(0.95, 0.65, textstr, fontsize=10)

# Save Figure
plt.savefig("Pyber bubble plot.png")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_total_fare = (urban_data["fare"].sum())*100
# print(urban_total_fare)
suburban_total_fare = suburban_data["fare"].sum()*100
rural_total_fare = rural_data["fare"].sum()*100

#driver_percentages = full_df['driver_count'].first()
 

# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
values = [urban_total_fare, suburban_total_fare, rural_total_fare]

colors = ["lightcoral", "lightskyblue", "gold"]
explode = [.08, 0, 0]
plt.pie(values, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=290)


# Save Figure
plt.savefig("total_fares.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_rides = urban_data["ride_id"].count()
suburban_rides = suburban_data["ride_id"].count()
rural_rides = rural_data["ride_id"].count()


# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
values = [urban_rides, suburban_rides, rural_rides]
colors = ["lightcoral", "lightskyblue", "gold"]
explode = [.08, 0, 0]
plt.pie(values, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=250)


# Save Figure
plt.savefig("total_rides.png")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Separating information for different types
urban_driver_data = city_data.loc[city_data['type']=='Urban']
suburban_driver_data = city_data.loc[city_data['type']=='Suburban']
rural_driver_data = city_data.loc[city_data['type']=='Rural']

# Calculate Driver Percents
urban_total_drivers = urban_driver_data["driver_count"].sum()
suburban_total_drivers = suburban_driver_data["driver_count"].sum()
rural_total_drivers = rural_driver_data["driver_count"].sum()

# Build Pie Charts
labels = ["Urban", "Suburban", "Rural"]
values = [urban_total_drivers, suburban_total_drivers, rural_total_drivers]
colors = ["lightcoral", "lightskyblue", "gold"]
explode = [.08, 0, 0]
plt.pie(values, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=250)

# Save Figure
plt.savefig("total_driver.png")

# Show Figure
plt.show()
```
